```python
import pandas as pd
import numpy as np
```

```python
 # Cargar datos
df = pd.read_csv("../data/restore/comentarios_app_v2.csv", encoding="utf-8")
print("\n Cantidad de registros: ", len(df))
df.head()
```

### DESCRIPCION DEL DATASET
- reviewId: Identificador único de la reseña en Google Play (UUID generado por Google).
- userName: Nombre del usuario que publicó la reseña.
- userImage: URL de la imagen del usuario que publicó la reseña.
- content: Contenido de la reseña.
- score: Puntuación dada por el usuario (1-5).
- thumbsUpCount: Cantidad de votos positivos para la reseña.
- reviewCreatedVersion: Versión de la app en la que se creó la reseña.
- at: Fecha y hora en que se publicó la reseña.
- replyContent: Contenido de la respuesta a la reseña (si hay).
- repliedAt: Fecha y hora en que se respondió la reseña (si hay).
- appVersion: Versión de la app en la que se creó la reseña.
- app_id (extra): Identificador único de la app en Google Play.

```python
# Mostrar resumen de datos faltantes
faltantes = df.isnull().sum().sort_values(ascending=False)
total = len(df)

print("Cantidad de valores faltantes por columna:")
print(faltantes)

# Mostrar también en porcentaje
print("\n Porcentaje de valores faltantes por columna:")
print((faltantes / total * 100).round(2))
```
