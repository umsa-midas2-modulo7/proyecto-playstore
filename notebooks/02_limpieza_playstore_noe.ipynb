```python
import re
import string
import unicodedata
import pandas as pd
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.stem.snowball import SpanishStemmer
from spellchecker import SpellChecker

# ──────────────────────────────
# 1️⃣ DESCARGAR STOPWORDS
nltk.download("stopwords")
stop_words = set(stopwords.words("spanish"))

# ──────────────────────────────
# 2️⃣ CARGAR MODELO SPACY EN ESPAÑOL
nlp = spacy.load("es_core_news_sm")
stemmer = SpanishStemmer()

# ──────────────────────────────
# 3️⃣ INICIALIZAR SPELLCHECKER
spell = SpellChecker(language="es")

# ──────────────────────────────
# 4️⃣ LIMPIEZA BÁSICA (VECTORIZADA)
df["content"] = df["content"].astype(str)  # Evita NaN
df["content_limpio"] = (
    df["content"]
    .str.lower()
    .str.replace(r"http\S+|www\S+", "", regex=True)
    .str.replace(r"\d+", "", regex=True)
    .str.replace(f"[{string.punctuation}]", "", regex=True)
    .apply(lambda texto: " ".join([p for p in texto.split() if p not in stop_words]))
)

# ──────────────────────────────
# 4.1️⃣ ELIMINAR PALABRAS DE LONGITUD 2
def eliminar_palabras_cortas(texto, longitud=2):
    return " ".join([p for p in texto.split() if len(p) != longitud])

df["content_limpio"] = df["content_limpio"].map(lambda x: eliminar_palabras_cortas(x, longitud=2))

# ──────────────────────────────
# 5️⃣ CORRECCIÓN ORTOGRÁFICA EN BATCH (MUY OPTIMIZADO)
print("⚡ Generando diccionario de correcciones ortográficas...")
palabras_unicas = set(" ".join(df["content_limpio"]).split())

# Corrección solo una vez por palabra única
correcciones = {}
for p in palabras_unicas:
    if p not in stop_words:
        correc = spell.correction(p)
        correcciones[p] = correc if correc is not None else p
    else:
        correcciones[p] = p

print(f"✅ Diccionario creado: {len(correcciones)} palabras únicas.")

def corregir_texto(texto):
    return " ".join([correcciones.get(p, p) for p in texto.split()])

df["content_limpio_corregido"] = df["content_limpio"].map(corregir_texto)

# ──────────────────────────────
# 6️⃣ QUITAR ACENTOS (VECTORIZADO)
def quitar_acentos_vector(textos):
    return [
        "".join(c for c in unicodedata.normalize("NFD", texto) if unicodedata.category(c) != "Mn")
        for texto in textos
    ]

df["content_limpio_corregido"] = quitar_acentos_vector(df["content_limpio_corregido"].tolist())

# ──────────────────────────────
# 7️⃣ LEMATIZACIÓN EN BATCH CON SPACY (RÁPIDA)
print("⚡ Iniciando lematización con spaCy...")
def lematizar_sin_stemmer(textos):
    return [
        " ".join([token.lemma_ for token in doc if len(token.lemma_) > 2 and not token.is_stop])
        for doc in nlp.pipe(textos, batch_size=100, n_process=1)
    ]

df["content_normalizado"] = lematizar_sin_stemmer(df["content_limpio_corregido"].tolist())
print("✅ Lematización completada.")

# ──────────────────────────────
# 8️⃣ TOKENIZACIÓN RÁPIDA
df["tokens"] = df["content_normalizado"].str.split()

# ──────────────────────────────
# ✅ RESULTADO FINAL
df[["content", "content_limpio", "content_limpio_corregido", "content_normalizado", "tokens"]].head()
```

```python

import re
import string
import unicodedata
import pandas as pd
import nltk
import spacy
from nltk.corpus import stopwords
from nltk.stem.snowball import SpanishStemmer
from spellchecker import SpellChecker

# 1️⃣ Descargar stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("spanish"))

# 2️⃣ Cargar modelo spaCy en español
nlp = spacy.load("es_core_news_sm")
stemmer = SpanishStemmer()

# 3️⃣ Inicializar corrector ortográfico automático
spell = SpellChecker(language="es")

# 4️⃣ Función para limpieza básica
def limpiar_texto(texto):
    if pd.isnull(texto):
        return ""
    texto = texto.lower()
    texto = re.sub(r"http\S+|www\S+", "", texto)
    texto = re.sub(r"\d+", "", texto)
    texto = texto.translate(str.maketrans("", "", string.punctuation))
    palabras = texto.split()
    palabras = [p for p in palabras if p not in stop_words]
    return " ".join(palabras)

df["content_limpio"] = df["content"].apply(limpiar_texto)

# 5️⃣ Corregir errores ortográficos de forma automática (más rápida)
def corregir_textos(textos):
    textos_corregidos = []
    for texto in textos:
        palabras = texto.split()
        corregidas = [
            spell.correction(p) if p not in stop_words and spell.correction(p) is not None else p
            for p in palabras
        ]
        textos_corregidos.append(" ".join(corregidas))
    return textos_corregidos

df["content_limpio_corregido"] = corregir_textos(df["content_limpio"].tolist())

# 6️⃣ Quitar acentos
def quitar_acentos(texto):
    return ''.join(c for c in unicodedata.normalize('NFD', texto) if unicodedata.category(c) != 'Mn')

df["content_limpio_corregido"] = df["content_limpio_corregido"].apply(quitar_acentos)

def lematizar_sin_stemmer(textos):
    resultados = []
    for doc in nlp.pipe(textos, batch_size=100, n_process=2):
        lemas = [token.lemma_ for token in doc if len(token.lemma_) > 2 and not token.is_stop]
        resultados.append(" ".join(lemas))
    return resultados

df["content_normalizado"] = lematizar_sin_stemmer(df["content_limpio_corregido"].tolist())
df["tokens"] = df["content_normalizado"].apply(lambda x: x.split())


# ✅ Mostrar ejemplo
df[["content", "content_limpio", "content_limpio_corregido", "content_normalizado", "tokens"]].head()


```
